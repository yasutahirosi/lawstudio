```python
#Import Module.
import saspy
from sklearn import datasets
from IPython.display import display, HTML

#Open python default datasets Iris.
iris = datasets.load_iris(as_frame=True)['data']

#Set up connection to sas.
sas=saspy.SASsession()
#Turn pandas dataframe to sas7bdat and create table iris_sas in work library.
iris_sas=sas.dataframe2sasdata(iris,table="iris_sas")

#A simple sample code to show how to submit sascode
stats=sas.submit("proc univariate data=work.iris_sas;run;")
sas.HTML(stats["LST"])
```
